# Print with Format

6: This is the minimum width of the field. It means that the output will be at least 6 characters wide. If the formatted number is shorter than 6 characters, it will be padded with spaces on the left to make up the width. If the number is wider, it will not be truncated.

.2 specifies that there should be two digits after the decimal point.

.0 specifies that there should be zero digits after the decimal point. Essentially, this means the number will be rounded to the nearest integer.


f: This stands for floating-point number. It means the value should be formatted as a floating-point number.

```python
x = 543.12
print ("x after formatting:", "%6.2f" %x)
print ("Length after formatting:", len("%6.2f" %x))
```

```python
x = 543.12
print ("x after formatting:", "%8.2f" %x)
print ("Length after formatting:", len("%8.2f" %x))
```

```python
x = 543.12
print ("x after formatting:", "%8.4f" %x)
print ("Length after formatting:", len("%8.4f" %x))
```

```python
x = 543.12
print ("x after formatting:", "%1.4f" %x)
print ("Length after formatting:", len("%1.4f" %x))
```

```python
x = 543.12
print ("x after formatting:", "%8.5f" %x)
print ("Length after formatting:", len("%8.5f" %x))
```

the binary number 1010 (which is 10 in decimal) is represented as 12 in octal:

```python
# hexadecimal: The decimal number 255 is FF in hexadecimal.
# x: unsigned hexadecimal in lowercase
# X: unsigned hexadecimal in uppercase

number = 255
print ('number = %1x' %number)
print ('number = %1X' %number)
```

```python
# e: FLoating point with exponential format in lowercase
# E: FLoating point with exponential format in uppercase

number = 123456.789
print ('number = %1e' %number)
print ('number = %1EX' %number)
```

```python
""" The f and F(format floating-point numbers in decimal form). 
The difference between them is in how they represent special values like "nan" and "inf".
"""
nan_value = float('nan')
print ('number = %1f' %nan_value)
print ('number = %1F' %nan_value)
```

```python
#c: Formats integers as characters using the corresponding ASCII value.
ascii_value = 65
print ('number = %1c' %ascii_value)
```

# Try/Except Statements

```python
def divide_numbers(num1, num2):
    try:
        # Attempt to divide the numbers
        result = num1 / num2
        print(f"Result: {result}")
    except ZeroDivisionError:
        # Handle division by zero error
        print("Error: Cannot divide by zero!")
    except TypeError as e:
        # Handle type errors (e.g., passing non-numeric values)
        print(f"Error: {e}")
    finally:
        # This block will always execute
        print("Operation completed.")

# Example usage
divide_numbers(10, 2)  # Successful division
divide_numbers(10, 0)  # Division by zero
divide_numbers(10, 'a')  # Type error
```

```python
try:
    x = 1
    y = 10/x
    print ("The result is: ", y)
except ZeroDivisionError:
    print("Error: Division by zero")
except ValueError:
    print("Error: Invalid input, please enter a valid number")
else:
    print("No exceptions occured")
finally:
    print ("Excution complete")
print("Program finished")
```
